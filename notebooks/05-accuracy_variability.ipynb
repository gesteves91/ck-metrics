# Accuracy and Variability

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
sns.set_style("white")
```

```python
x = pd.read_csv('../experiments/accvar.csv', index_col=0)
```

#### Accuracy

```python
df = x.sort_values('acc',ascending=False)
```

```python
df['feats'] = np.arange(len(df))
```

```python
plt.figure(figsize=(2,3)) # this creates a figure 8 inch wide, 4 inch high
ax = sns.lineplot(x="feats", y="acc", color='blue',data=df)
ax.set_xticks([])
ax.set(ylabel='AUC (mean)', xlabel='features (by AUC mean)')
plt.savefig('../reports/acc.eps', bbox_inches='tight', dpi=800)
plt.show()
```

#### Variability

```python
df2 = x.sort_values('mad',ascending=False)
```

```python
df2['feats'] = np.arange(len(df))
```

```python
plt.figure(figsize=(2,3)) # this creates a figure 8 inch wide, 4 inch high
ax = sns.lineplot(x="feats", y="mad", color='red',data=df2)
ax.set_xticks([])
ax.set(ylabel='AUC (mean)', xlabel='features (by AUC mean)')
plt.savefig('../reports/var.eps', bbox_inches='tight', dpi=800)
plt.show()
```
